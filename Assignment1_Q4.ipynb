```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab    
import json
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
import powerlaw
```

```python
new_review = pd.read_csv('new_review.csv')
new_business = pd.read_csv('new_business.csv')
```

```python
df_La= new_business[new_business.city=='Las Vegas']['review_count']
```

```python
fit = powerlaw.Fit(df_La)
fit.power_law.alpha
```

```python
fit.power_law.xmin
```

### Plot of the probability density function with the empirical data

```python
powerlaw.plot_pdf(df_La, color = 'b')
powerlaw.plot_pdf(df_La, linear_bins = True, color = 'r')
```

### Visualizing just the portion of the data using for fitting to the distribution

```python
fit = powerlaw.Fit(df_La)
fig, axs = plt.subplots(1,2, figsize=(12, 4), facecolor='w', edgecolor='b')
fig.subplots_adjust(hspace = .3, wspace=.2)
axs = axs.ravel()

fit.plot_ccdf(color = 'b', linewidth = 2,ax = axs[0])
fit.power_law.plot_ccdf(color = 'b', linestyle = '--', ax = axs[0])
axs[0].set_title('CCDF of empirical data and fitted power law')

fit.plot_pdf(color = 'b', linewidth = 2,ax = axs[1])
fit.power_law.plot_pdf(color = 'b', linestyle = '--', ax = axs[1])
axs[1].set_title('PDF of empirical data and fitted power law')
```
