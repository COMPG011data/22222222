```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab    
import json
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
```

```python
new_business = pd.read_csv('new_business.csv')
business_review= new_business[new_business.city=='Las Vegas']['review_count']
```

```python
par=business_review.value_counts().sort_index()
freq=np.cumsum(par)/(sum(par)+1)
freq=1-freq

plt.plot(np.log(par.index),np.log(freq),"ro",linewidth=1)
alpha = len(par.index)/sum(np.log(par.index/min(par.index)))
```
